```python
import csv
import pandas as pd
import numpy as np
import os
import re
```

```python
path_name = 'Antioxidant'
```

```python
# path
path_dir = os.path.dirname(os.path.dirname(os.getcwd()))

path_merge = path_dir + '\\new_data\\' + path_name+r'\merge_feature.csv'

path_data = path_dir + '\\new_data\\' + path_name+r'\merge_data.csv'

path_new_positive = path_dir + '\\new_data\\' + path_name+ r'\merge\positive.csv'
path_new_negative = path_dir + '\\new_data\\' + path_name + r'\merge\negitive.csv'
```

```python
with open(path_merge,'r') as f:
    features = f.readlines()
features = features[0].replace('\n','')
features = features.split(sep = ',')
```

```python
# 将正类和负类数据抽取成list
def extract_data(path):
    lines_list =[]
    with open(path, 'r') as f:
        for line in f.readlines():
            line = line.replace('\n','')
            lines_list.append(line)
    return lines_list
```

```python
# 抽取正负样本数据
positive_data = extract_data(path_new_positive)
negative_data = extract_data(path_new_negative)
```

```python
def get_frequency(feature, data, pat= '[A-Z]'):
    # 获取某一个特征在某一样本中出现的次数
    feature_list = []
    pattern = r''
    for s in feature:
        if s!='*':
            pattern += s
        else:
            pattern += pat   # 修改
    for d in data:
        flag = re.findall(pattern, d)
        if flag != None:
            freq = len(flag)
        else:
            freq = 0
        feature_list.append(freq)
    return feature_list
```

```python
new_features = ['label']
new_features.extend(features)
po = (len(positive_data))*[1]
ne = (len(negative_data))*[0]
po.extend(ne)
```

```python
new_data_list = []
new_data_list.append(po)
for feature in features:
    positive_list = get_frequency(feature, positive_data, '[a-m]')
    negitive_list = get_frequency(feature, negative_data, '[a-m]')
    
    positive_list.extend(negitive_list)
    new_data_list.append(positive_list)
```

```python
data_dict = dict(zip(new_features, new_data_list))
```

```python
final_data = pd.DataFrame(data_dict)
```
